## ¿Qué es un paradigma de programación?
Se define paradigma de programacion como un conjunto de principios y directrices que define un enfoque particular para diseñar, estructurar y escribir un código. Se divide principalmente en paradigmas imperativos y paradigmas declarativos.
Los paradigmas imperativos tienen su código relacionado a la arquitectura de la máquina y su secuencia de pasos (instrucciones) cambian el estado delprograma. Dentro de los imperativos estan:
-Procedimental: Enfatiza aplicar procedimientos de forma similar al funcionamiento de la máquina y se puede reutilizar el codigo.
-Orientado a objetos: Se usa una clase, las cuales son una suerte de molde o matriz que tiene el fin de crear un objeto con atributos y funcionalidades. Tiene mayor enfasis en los datos que en los procedimientos.

## ¿En qué se basa la programación orientada a objetos?
Es un tipo de paradigma imperativo, el cual usa una clase, las cuales son una suerte de molde o matriz que tiene el fin de crear un objeto con atributos y funcionalidades. Tiene mayor enfasis en los datos que en los procedimientos. Es de los paradigmas mas utilizados debido a que es adecuado para la creacion de aplicaciones simples y complejas, se puede reutilizar el codigo, se puede incorporar fácilmente nuevas funciones al código existente y reduce los costos de produccion y mantenimiento.


## ¿Cuál es la diferencia entre recursividad e iteración, y cómo se relaciona esto con la notación big 𝑂?
La dieferencia entre recursividad e iteracion es en como se estructura el codigo, en recursividad la funcion se llama a si misma de forma directa o indirecta, en la iteracion el mismo procedimiento se repite varias veces usando bucles for o while. La notacion Big O es el lenguaje que se utiliza para ver el tiempo que tarda en ejecutarse un algoritmo, se relaciona con la recursividad e iteracion debido a que estas tienen distinto tiempo de ejecucion. 

## explicar la diferencia de rendimiento entre 𝑂(1) y 𝑂(𝑛)
Los algoritmos de tiempo constante O(1) ejecutan en un tiempo constante y el número de operaciones que debe realizar el algoritmo para completar una tarea determinada es independiente del tamaño de la entrada, es el algoritmo mas rapido. Los algoritmos de tiempo constante O(n) ejecutan en un tiempo lineal, crece directamente proporcional al tamaño de sus entradas, es el tercer algoritmo mas rapido.


## ¿Cómo se calcula el orden en un programa que funciona por etapas?
Para calcular el orden en un programa que funciona por etapas se debe calcular el orden de cada etapa por separado e identificar la etapa que tiene mayor orden, ese sera el orden del programa.

```python
#Dos soluciones
#Solucion recursiva
class PCBrecursivo:
    def __init__(self,N,M):
        self.N = N
        self.M = M
    def caminos(self,ancho,largo):
        #Caso base si solo queda un movimiento para llegar a B
        if largo == 1 or ancho == 1:
            return 1
        #No hay mas que recorrer
        if largo < 0 or ancho < 0:
            return 0
        #Aun tiene por recorrer
        else:
            return self.caminos(ancho - 1, largo) + self.caminos(ancho, largo - 1)
a = PCBrecursivo(10,4)
print("Caminos com solucion recursiva: ",a.caminos(5,6))
#Solucion iterativa
class PCBiterativo:
    def __init__(self,N,M):
      self.N = N
      self.M = M
      self.arreglo = [[0 for _ in range(M)] for _ in range(N)]
    def caminos(self):
       for j in range(self.M):
          self.arreglo[0][j] = 1
       for i in range(self.N):
          self.arreglo[i][0] = 1
       for i in range(1,self.N):
          for j in range(1,self.M):
             self.arreglo[i][j] = self.arreglo[i-1][j] + self.arreglo[i][j-1]
       return self.arreglo[self.N-1][self.M-1]
b = PCBiterativo(5,6)
print("Caminos con solucion iterativa",b.caminos())           
```

```python
#parte 1
import math
pcbrecursivo = []
pcbiterativo = []
#Calcular tiempo de ejecucion
from time import perf_counter
for i in range(1, 100):
    #Tiempo ejecución función pcb2
    tiempo_inicio = perf_counter()
    PCBrecursivo(50,i)
    tiempo_total = perf_counter() - tiempo_inicio
    pcbrecursivo.append(tiempo_total)
    #Tiempo ejecucion funcion pcb1
    tiempo_inicio = perf_counter()
    PCBiterativo(50,i)
    tiempo_total = perf_counter() - tiempo_inicio
    pcbiterativo.append(tiempo_total)   
```

```python
plt.plot(pcbrecursivo, label=r"recursion")
plt.plot(pcbiterativo, label=r"iteracion")
plt.grid(True)
plt.xlabel('Ejecuciones')
plt.ylabel('segundos')
plt.legend()
plt.show()
```
